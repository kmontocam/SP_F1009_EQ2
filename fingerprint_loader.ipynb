```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from scipy.io import wavfile
from fingerprinter import fingerprinter
from compressor import compressor
```

```python
directory = r'/Users/kmontocam/Documents/GitHub/tracks'
MUSIC_FOLDER = os.listdir(directory)
tracks = [file for file in MUSIC_FOLDER if (file[-4:] == '.wav')]

for track in tracks:

    song_freq, song = wavfile.read(os.path.join(directory, track))

    compressed_freq, song_compressed = compressor(song_freq, song)

    freq_times, freq_highs = fingerprinter(song_freq, song_compressed)

    plt.scatter(freq_times, freq_highs)
    plt.show()

    #df = pd.DataFrame({'time': freq_times, 'hz': freq_highs})
    #df.columns.name = track
    #df.to_csv(f'{track[:-4]}.csv', index = False)
```
